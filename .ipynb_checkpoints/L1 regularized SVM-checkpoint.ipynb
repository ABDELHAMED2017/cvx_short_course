```python
from __future__ import division
import numpy as np
import scipy.sparse as sp
from cvxpy import *


np.random.seed(1)
n = 20
m = 1000
TEST = m
DENSITY = 0.2
beta_true = np.random.randn(n,1)
idxs = np.random.choice(range(n), int((1-DENSITY)*n), replace=False)
print idxs
for idx in idxs:
    beta_true[idx] = 0
# sigma = np.sqrt(n)
offset = 0#np.sqrt(n)
X = np.random.normal(0, 5, size=(m,n))
sigma = 45#np.sqrt(n)
Y = np.sign(X.dot(beta_true) + offset + np.random.normal(0,sigma,size=(m,1)))

X_test = np.random.normal(0, 5, size=(TEST,n))
Y_test = np.sign(X_test.dot(beta_true) + offset + np.random.normal(0,sigma,size=(TEST,1)))


beta = Variable(n)
v = Variable()
loss = sum_entries(pos(1 - mul_elemwise(Y, X*beta - v)))
reg = norm(beta, 1)
gamma = Parameter(sign="positive")
prob = Problem(Minimize(loss/m + gamma*reg))

TRIALS = 50
train_error = np.zeros(TRIALS)
test_error = np.zeros(TRIALS)
gamma_vals = np.logspace(-4, 0, TRIALS)
for i in range(TRIALS):
    gamma.value = gamma_vals[i]
    prob.solve()
#     print i, (abs(beta).value > 1e-6).sum()
    train_error[i] = (np.sign(X.dot(beta_true) + offset) != np.sign(X.dot(beta.value) - v.value)).sum()/m
    test_error[i] = (np.sign(X_test.dot(beta_true) + offset) != np.sign(X_test.dot(beta.value) - v.value)).sum()/TEST
                           
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(gamma_vals, train_error, label="Train error")
plt.plot(gamma_vals, test_error, label="Test error")
plt.xscale('log')
plt.legend(loc='upper left')
plt.xlabel(r"$\gamma$", fontsize=16)
# plt.savefig('L1_SVM')
plt.show()
print train_error.min()
print test_error.min()
print test_error[0]
```

```python
from __future__ import division
import numpy as np
import scipy.sparse as sp
from cvxpy import *


np.random.seed(1)
n = 20
m = 1000
TEST = m
DENSITY = 0.2
beta_true = np.random.randn(n,1)
idxs = np.random.choice(range(n), int((1-DENSITY)*n), replace=False)
print idxs
for idx in idxs:
    beta_true[idx] = 0
# sigma = np.sqrt(n)
offset = 0#np.sqrt(n)
X = np.random.normal(0, 5, size=(m,n))
sigma = 45#np.sqrt(n)
Y = np.sign(X.dot(beta_true) + offset + np.random.normal(0,sigma,size=(m,1)))

X_test = np.random.normal(0, 5, size=(TEST,n))
Y_test = np.sign(X_test.dot(beta_true) + offset + np.random.normal(0,sigma,size=(TEST,1)))


beta = Variable(n)
v = Variable()
loss = sum_entries(pos(1 - mul_elemwise(Y, X*beta - v)))
reg = norm(beta, 1)
gamma = Parameter(sign="positive")
prob = Problem(Minimize(loss/m + gamma*reg))

TRIALS = 100
train_error = np.zeros(TRIALS)
test_error = np.zeros(TRIALS)
gamma_vals = np.logspace(-2, 0, TRIALS)
beta_vals = []
for i in range(TRIALS):
    gamma.value = gamma_vals[i]
    prob.solve()
#     print i, (abs(beta).value > 1e-6).sum()
    train_error[i] = (np.sign(X.dot(beta_true) + offset) != np.sign(X.dot(beta.value) - v.value)).sum()/m
    test_error[i] = (np.sign(X_test.dot(beta_true) + offset) != np.sign(X_test.dot(beta.value) - v.value)).sum()/TEST
    beta_vals.append(beta.value)                     
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(gamma_vals, train_error, label="Train error")
plt.plot(gamma_vals, test_error, label="Test error")
plt.xscale('log')
plt.legend(loc='upper left')
plt.xlabel(r"$\lambda$", fontsize=16)
plt.savefig('L1_SVM')
plt.show()
print train_error.min()
print test_error.min()
print test_error[0]
```
