```python
import numpy as np
from cvxpy import *
np.random.seed(1)
SAMPLES = 100

n = 50
w_true = np.random.randn(n, 1)
A = np.random.randn(SAMPLES, n)
A[:,-1] = 1
v = np.random.normal(size=(SAMPLES,1))
b = A.dot(w_true) + v
w = Variable(n)
r = b - A*w
tau = Parameter()
cost = sum_entries(0.5*abs(r) + (tau - 0.5)*r)
prob = Problem(Minimize(cost))

# Plotting
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.subplots(1,3)
tau_vals = [0.1, 0.5, 0.9]
for i, tau_val in enumerate(tau_vals):
    tau.value = tau_val
    prob.solve()
    print norm(w[:-1]).value
    plt.subplot(1,3,i+1)
    plt.hist(r.value)
    plt.title(r'$\tau = %.1f$' % tau_val)
    plt.xlabel('Residuals')
plt.show()
```

```python
from __future__ import division
np.random.seed(1)
TRAIN_LEN = 400
SKIP_LEN = 100
TEST_LEN = 100
TOTAL_LEN = TRAIN_LEN + SKIP_LEN + TEST_LEN
m = 10
x0 = np.random.randn(m)
x = np.zeros(TOTAL_LEN) 
x[:m] = x0 #+ np.random.normal()
for i in range(m+1, TOTAL_LEN):
    x[i] = 0.5*log_sum_exp(x[i-m:i]).value + np.random.normal()
x_true = x
# x = x + np.random.normal(TOTAL_LEN)
w = Variable(m)
v = Variable()
tau = Parameter()
error = 0
for i in range(SKIP_LEN, TRAIN_LEN + SKIP_LEN):
    r = x[i] - (w.T*x[i-m:i] + v)
    error += 0.5*abs(r) + (tau - 0.5)*r
prob = Problem(Minimize(error))

tau_vals = [0.9, 0.5, 0.1]
# Prediction:
pred = np.zeros((len(tau_vals), TOTAL_LEN))
r_vals = np.zeros((len(tau_vals), TOTAL_LEN))
for k, tau_val in enumerate(tau_vals):
    tau.value = tau_val
    prob.solve()
    pred[k,:m] = x0
    for i in range(SKIP_LEN, TOTAL_LEN):
        pred[k,i] = (x[i-m:i].T*w + v).value
        r_vals[k, i] = (x[i] - (x[i-m:i].T*w + v)).value

import matplotlib.pyplot as plt
%matplotlib inline
# plt.figure(figsize=(10,7))
plt.plot(range(0, TRAIN_LEN, 4), x[SKIP_LEN+1:-TEST_LEN:4],  'black', label=r'$x$')
colors = ['b', 'g', 'r']
for k, tau_val in enumerate(tau_vals):
    plt.plot(range(0, TRAIN_LEN, 4), pred[k,SKIP_LEN+1:-TEST_LEN:4], colors[k],  label=r"$\tau = %.1f$" % tau_val)
# plt.legend(loc='upper left')
plt.xlabel(r"$t$", fontsize=16)
plt.ylabel(r"$x_t$", fontsize=16)
plt.savefig('quantile_train')
plt.show()

plt.plot(range(TRAIN_LEN,TRAIN_LEN+TEST_LEN), x[-TEST_LEN:],  'black', label=r'$x$')
for k, tau_val in enumerate(tau_vals):
    plt.plot(range(TRAIN_LEN,TRAIN_LEN+TEST_LEN), pred[k,-TEST_LEN:], colors[k],  label=r"$\tau = %.1f$" % tau_val)
# plt.legend(loc='upper left')
plt.xlabel(r"$t$", fontsize=16)
plt.ylabel(r"$x_t$", fontsize=16)
plt.savefig('quantile_test')
plt.show()
print (pred[0, SKIP_LEN:-TEST_LEN] < x[SKIP_LEN:-TEST_LEN]).sum()/TRAIN_LEN
print (x[SKIP_LEN:-TEST_LEN] < pred[2,SKIP_LEN:-TEST_LEN]).sum()/TRAIN_LEN

print (pred[0,-TEST_LEN:] < x[-TEST_LEN:]).sum()/TEST_LEN
print (x[-TEST_LEN:] < pred[2,-TEST_LEN:]).sum()/TEST_LEN
```

```python
# from scipy.stats import cumfreq
for k, tau_val in enumerate(tau_vals):
    sorted=np.sort(r_vals[k,-TEST_LEN:] )
    yvals=np.arange(len(sorted))/float(len(sorted))
    plt.plot( sorted, yvals, colors[k], label=r"$\tau = %.1f$" % tau_val)

x = np.linspace(-4, 6, 100)
for val in [0.1, 0.5, 0.9]:
    plt.plot(x, len(x)*[val], 'k--')
plt.xlabel('Residual')
plt.ylabel('Cumulative density')
# plt.legend(loc='upper left')
plt.savefig('quantile_test_cdf')
plt.show()


# from scipy.stats import cumfreq
for k, tau_val in enumerate(tau_vals):
    sorted=np.sort(r_vals[k,SKIP_LEN:SKIP_LEN+TRAIN_LEN] )
    yvals=np.arange(len(sorted))/float(len(sorted))
    plt.plot( sorted, yvals, colors[k], label=r"$\tau = %.1f$" % tau_val)

x = np.linspace(-6, 6, 100)
for val in [0.1, 0.5, 0.9]:
    plt.plot(x, len(x)*[val], 'k--')
plt.xlabel('Residual')
plt.ylabel('Cumulative density')
# plt.legend(loc='upper left')
plt.savefig('quantile_train_cdf')
plt.show()
```
