```python
from cvxpy import *
import numpy as np
np.random.seed(1)
n = 30
r = np.random.randn(n, 1)
Sigma = np.random.randn(n, n)
Sigma = Sigma.T.dot(Sigma)

w = Variable(n)
gamma = Parameter(sign='positive')
ret = r.T*w 
risk = quad_form(w, Sigma)
prob = Problem(Maximize(ret - gamma*risk), 
               [sum_entries(w) == 1, 
                w >= 0])

# Plotting.
SAMPLES = 100
risk_data = np.zeros(SAMPLES)
ret_data = np.zeros(SAMPLES)
gamma_vals = np.logspace(-2, 3, num=SAMPLES)
for i in range(SAMPLES):
    gamma.value = gamma_vals[i]
    prob.solve()
    risk_data[i] = risk.value
    ret_data[i] = ret.value
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(risk_data, ret_data)
plt.xlabel('Standard deviation or risk')
plt.ylabel('Return')
plt.show()
```

```python
import numpy as np
import scipy.linalg

n = 100
m = 30
D = np.diag(np.random.uniform(size=n))
F = np.random.uniform(-.1, 1.1, size=(n,m))
Sigma_s = np.random.normal(size=(m,m))
Sigma_s = Sigma_s.T.dot(Sigma_s)
mu = np.random.uniform(size=(n,1))

Sigma_rt = np.real(scipy.linalg.sqrtm(Sigma_s))
D_rt = np.real(scipy.linalg.sqrtm(D))
B = 1
```

```python
from cvxpy import *

kappa = Parameter(sign='positive')
kappa.value = 0
h = Variable(n)
s = F.T*h
risk = norm(vstack(norm(Sigma_rt*s), norm(D_rt*h)))
exp_ret = mu.T*h - kappa*sum_entries(neg(h))
gamma = Parameter(sign='positive')

prob = Problem(Maximize(exp_ret - gamma*risk),
               [sum_entries(h) == B,
                h >= 0])

SAMPLES = 25
ret_data = np.zeros(SAMPLES)
risk_data = np.zeros(SAMPLES)
gamma_vals = np.logspace(-2, 3, num=SAMPLES)
for idx, val in enumerate(gamma_vals):
    gamma.value = val
    prob.solve()
    ret_data[idx] = exp_ret.value
    risk_data[idx] = risk.value
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(risk_data, ret_data)
plt.xlabel('Standard deviation or risk')
plt.ylabel('Return')
plt.show()
```

```python
prob = Problem(Maximize(exp_ret - gamma*risk),
               [sum_entries(h) == B, exp_ret >= 0])


SAMPLES = 25
NUM_KAPPAS = 5
kappa_vals =  np.logspace(0, 1, num=NUM_KAPPAS)
gamma_vals = np.logspace(-2, 3, num=SAMPLES)
ret_data = np.zeros((NUM_KAPPAS, SAMPLES))
risk_data = np.zeros((NUM_KAPPAS, SAMPLES))
for idx1, kval in enumerate(kappa_vals):
    kappa.value = kval
    for idx2, gval in enumerate(gamma_vals):
        gamma.value = gval
        prob.solve(solver=GUROBI)
        ret_data[idx1,idx2] = exp_ret.value
        risk_data[idx1,idx2] = risk.value
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
for idx1, kval in enumerate(kappa_vals):
    plt.plot(risk_data[idx1,:], ret_data[idx1,:], label="kappa = %.1f" % kval)
plt.xlabel('Standard deviation or risk')
plt.ylabel('Return')
plt.legend(loc='upper left')
plt.show()
```
