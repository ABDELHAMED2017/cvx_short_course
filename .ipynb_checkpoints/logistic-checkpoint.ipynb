```python
from __future__ import division
import numpy as np
import scipy.sparse as sp
from cvxpy import *

def pairs(X):
    m, n = X.shape
    k = n*(n+1)//2
    Z = np.zeros((m,k))
    count = 0
    for i in range(n):
        for j in range(i,n):
            Z[:,count] = X[:,i]*X[:,j]
            count += 1
    return Z
    
np.random.seed(1)
n = 10
k = n*(n+1)//2
m = 200
TEST = 100
sigma = 1.9 # 2
offset = 0
DENSITY = 1.0
beta_true = np.random.randn(n,1)
idxs = np.random.choice(range(n), int((1-DENSITY)*n), replace=False)
print idxs
for idx in idxs:
    beta_true[idx] = 0

X = np.random.binomial(1, 0.5, size=(m,n))
Y = np.sign(X.dot(beta_true) + np.random.normal(0,sigma,size=(m,1)))
Z = pairs(X)
X_test = np.random.binomial(1, 0.5, size=(TEST,n))
Y_test = np.sign(X_test.dot(beta_true) + np.random.normal(0,sigma,size=(TEST,1)))
Z_test = pairs(X_test)


beta = Variable(k)
v = Variable()
gamma = Parameter(sign="positive")
loss = 0
for i in range(m):
    loss += log_sum_exp(vstack(0, -Y[i]*(Z[i,:].T*beta - v)))
reg = norm(beta, 1)
prob = Problem(Minimize(loss/m + gamma*reg))

TRIALS = 100
train_error = np.zeros(TRIALS)
test_error = np.zeros(TRIALS)
gamma_vals = np.logspace(-4, 0, TRIALS)
for i in range(TRIALS):
    gamma.value = gamma_vals[i]
    prob.solve(solver=SCS)
#     print i, (abs(beta).value > 1e-6).sum()
    train_error[i] = (np.sign(X.dot(beta_true) + offset) != np.sign(Z.dot(beta.value) - v.value)).sum()/m
    test_error[i] = (np.sign(X_test.dot(beta_true) + offset) != np.sign(Z_test.dot(beta.value) - v.value)).sum()/TEST
                           
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(gamma_vals, train_error, label="Train error")
plt.plot(gamma_vals, test_error, label="Test error")
plt.xscale('log')
plt.legend(loc='upper left')
plt.xlabel(r"$\lambda$", fontsize=16)
# plt.savefig('L1_logistic')
plt.show()
print train_error.min()
print test_error.min()
print test_error[0]
```

```python
idx = np.argmin(test_error)
gamma.value = gamma_vals[idx]
prob.solve(solver=SCS)

P = np.zeros((n,n))
count = 0
for i in range(n):
    for j in range(i,n):
        P[i,j] = np.abs(beta.value[count])
        count += 1
# row_labels = list('ABCDE')
row_labels = range(1,n+1)
column_labels = range(1,n+1)

fig, ax = plt.subplots()#figsize=(15,5))
# data = D.value.A
# col_sums = data.sum(axis=0)
# data = data / col_sums[np.newaxis,:]
heatmap = ax.pcolor(P, cmap=plt.cm.Blues)

# put the major ticks at the middle of each cell
ax.set_xticks(np.arange(P.shape[1])+0.5, minor=False)
ax.set_yticks(np.arange(P.shape[0])+0.5, minor=False)

# # want a more natural, table-like display
ax.invert_yaxis()
ax.xaxis.tick_top()

ax.set_xticklabels(column_labels, minor=False)
ax.set_yticklabels(row_labels, minor=False)

plt.xlabel(r"$z_i$", fontsize=16)
ax.xaxis.set_label_position('top') 
plt.ylabel(r"$z_j$", fontsize=16)
# plt.savefig('logistic_feature_grid')
plt.show()
```

```python
from __future__ import division
import numpy as np
import scipy.sparse as sp
from cvxpy import *

def pairs(X):
    m, n = X.shape
    k = n*(n+1)//2
    Z = np.zeros((m,k))
    count = 0
    for i in range(n):
        for j in range(i,n):
            Z[:,count] = X[:,i]*X[:,j]
            count += 1
    return Z
    
np.random.seed(1)
n = 20
k = n*(n+1)//2
m = 200
TEST = 100
sigma = 1
offset = 0
DENSITY = 0.2
beta_true = np.random.randn(n,1)
idxs = np.random.choice(range(n), int((1-DENSITY)*n), replace=False)
for idx in idxs:
    beta_true[idx] = 0

X = np.random.binomial(1, 0.5, size=(m,n))
Y = np.sign(X.dot(beta_true) + np.random.normal(0,sigma,size=(m,1)))
Z = pairs(X)
X_test = np.random.binomial(1, 0.5, size=(TEST,n))
Y_test = np.sign(X_test.dot(beta_true) + np.random.normal(0,sigma,size=(TEST,1)))
Z_test = pairs(X_test)


beta = Variable(k)
v = Variable()
gamma = Parameter(sign="positive")
loss = 0
for i in range(m):
    loss += log_sum_exp(vstack(0, -Y[i]*(Z[i,:].T*beta - v)))
reg = norm(beta, 1)
prob = Problem(Minimize(loss/m + gamma*reg))

TRIALS = 100
train_error = np.zeros(TRIALS)
test_error = np.zeros(TRIALS)
gamma_vals = np.logspace(-3, -1, TRIALS)
beta_vals = []
for i in range(TRIALS):
    gamma.value = gamma_vals[i]
    prob.solve(solver=SCS)
#     print i, (abs(beta).value > 1e-6).sum()
    train_error[i] = (np.sign(X.dot(beta_true) + offset) != np.sign(Z.dot(beta.value) - v.value)).sum()/m
    test_error[i] = (np.sign(X_test.dot(beta_true) + offset) != np.sign(Z_test.dot(beta.value) - v.value)).sum()/TEST
    beta_vals.append(beta.value)
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(gamma_vals, train_error, label="Train error")
plt.plot(gamma_vals, test_error, label="Test error")
plt.xscale('log')
plt.legend(loc='upper right')
plt.xlabel(r"$\lambda$", fontsize=16)
plt.savefig('L1_logistic')
plt.show()
print train_error.min()
print test_error.min()
print test_error[0]
```
