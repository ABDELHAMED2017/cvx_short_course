```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import scipy.special
x = np.linspace(-10,10,100)
plt.figure(figsize=(10,2))
# plt.subplots(1,3)
plt.subplot(1,3,1)
plt.plot(x, np.square(x + 1), 'black')
plt.axis('off')

plt.subplot(1,3,2)
plt.plot(x, -(7*np.abs(x) + np.square(x)), 'black')
plt.axis('off')

plt.subplot(1,3,3)
plt.plot(x, 2*x + 1,  'black')
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import scipy.special
x = np.linspace(-1,1,100)
plt.figure(figsize=(7,2))
# plt.subplots(1,3)
plt.subplot(1,2,1)
plt.plot(x, 0.5*np.square(x), 'b')
plt.axis('off')

plt.subplot(1,2,2)
plt.plot(x, scipy.special.huber(.5, x), 'g')
plt.axis('off')
plt.savefig('huber_vs_square')
plt.show()
```
