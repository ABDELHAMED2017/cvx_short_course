```python
from cvxpy import *
import numpy as np
np.random.seed(1)
n = 10
r = np.abs(np.random.randn(n, 1))
Sigma = np.random.randn(n, n)
Sigma = Sigma.T.dot(Sigma)

w = Variable(n)
gamma = Parameter(sign='positive')
ret = r.T*w 
risk = quad_form(w, Sigma)
prob = Problem(Maximize(ret - gamma*risk), 
               [sum_entries(w) == 1, 
                w >= 0])
```

```python
# Plotting.
SAMPLES = 100
risk_data = np.zeros(SAMPLES)
ret_data = np.zeros(SAMPLES)
gamma_vals = np.logspace(-2, 3, num=SAMPLES)
for i in range(SAMPLES):
    gamma.value = gamma_vals[i]
    prob.solve()
    risk_data[i] = sqrt(risk).value
    ret_data[i] = ret.value
import matplotlib.pyplot as plt
%matplotlib inline

markers_on = [29, 40]
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(risk_data, ret_data, 'g-')
for marker in markers_on:
    plt.plot(risk_data[marker], ret_data[marker], 'bs')
    ax.annotate(r"$\gamma = %.2f$" % gamma_vals[marker], xy=(risk_data[marker]+.08, ret_data[marker]-.03), textcoords='offset points')
for i in range(n):
    plt.plot(sqrt(Sigma[i,i]).value, r[i], 'ro')
plt.xlabel('Standard deviation')
plt.ylabel('Return')
# plt.show()
# plt.savefig('long_only')
```

```python
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
%matplotlib inline
plt.figure()
for midx, idx in enumerate(markers_on):
    gamma.value = gamma_vals[idx]
    prob.solve()
    x = np.linspace(-2, 5, 1000)
    plt.plot(x, mlab.normpdf(x, ret.value, risk.value), label=r"$\gamma = %.2f$" % gamma.value)

plt.xlabel('Return')
plt.ylabel('Density')
plt.legend(loc='upper right')
# plt.show()
# plt.savefig('returns')
```

```python
# Simulate chosen points.
import matplotlib.pyplot as plt
%matplotlib inline
SAMPLES = 100
plt.figure()
plt.subplots(1,2)
for midx, idx in enumerate(markers_on):
    plt.subplot(1, 2, midx+1)
    gamma.value = gamma_vals[idx]
    prob.solve()
    ret_data = np.zeros(SAMPLES)
    for i in range(SAMPLES):
        r_val = np.random.multivariate_normal(r.flatten(), Sigma)
        ret_data[i] = (w.T*r_val).value
    plt.hist(ret_data)
#     plt.plot(risk_data[markers_on], ret_data[markers_on], 'ro')
    plt.xlabel('Return')
    plt.ylabel('Samples')
    plt.title("Risk = %f" % risk.value)
plt.show()
```

```python
# Leverage limit.
w = Variable(n)
gamma = Parameter(sign='positive')
Lmax = Parameter()
ret = r.T*w 
risk = quad_form(w, Sigma)
prob = Problem(Maximize(ret - gamma*risk), 
               [sum_entries(w) == 1, 
                norm(w, 1) <= Lmax])
# Plotting.
HOLDING_IDX = 20
L_vals = [1, 2, 4]
SAMPLES = 100
risk_data = np.zeros((len(L_vals), SAMPLES))
ret_data = np.zeros((len(L_vals), SAMPLES))
gamma_vals = np.logspace(-2, 3, num=SAMPLES)
w_vals = []
for k, L_val in enumerate(L_vals):
    for i in range(SAMPLES):
        Lmax.value = L_val
        gamma.value = gamma_vals[i]
        prob.solve()
#         print prob.value
        risk_data[k, i] = sqrt(risk).value
        ret_data[k, i] = ret.value
        
        
for k, L_val in enumerate(L_vals):
    prob = Problem(Maximize(ret), 
               [sum_entries(w) == 1, 
                norm(w, 1) <= Lmax,
                risk <= 2])
    Lmax.value = L_val
    prob.solve()
    w_vals.append( w.value )

    
import matplotlib.pyplot as plt
%matplotlib inline

# Lmax = infty has much larger risk and return than all others.
# Doesn't fit on plot.
for idx, L_val in enumerate(L_vals):
    plt.plot(risk_data[idx,:], ret_data[idx,:], label=r"$L^{\max}$ = %d" % L_val)
for w_val in w_vals:
    w.value = w_val
    plt.plot(sqrt(risk).value, ret.value, 'bs')
plt.xlabel('Standard deviation')
plt.ylabel('Return')
plt.legend(loc='lower right')
# plt.savefig('leverage_limits')
plt.show()

colors = ['b', 'g', 'r']
indices = np.argsort(r.flatten())
for idx, L_val in enumerate(L_vals):
     plt.bar(np.arange(1,n+1) + 0.25*idx - 0.375, w_vals[idx][indices], color=colors[idx], label=r"$L^{\max}$ = %d" % L_val, width = 0.25)
# plt.legend(loc='upper left')
plt.ylabel(r"$w_i$", fontsize=16)
plt.xlabel(r"$i$", fontsize=16)
plt.xlim([1-0.375, 10+.375])
plt.xticks(np.arange(1,n+1))
# plt.title(r"$w \; \mathrm{ for } \; \gamma = %.2f$" % gamma_vals[HOLDING_IDX], fontsize=16)
# plt.savefig('holdings')
plt.show()

```

```python
n = 3000
m = 50
np.random.seed(1)
r = np.abs(np.random.randn(n, 1))
Sigma = np.random.randn(m, m)
Sigma = Sigma.T.dot(Sigma)
D = np.diag(np.random.uniform(0, 0.9, size=n))
F = np.random.randn(m, n)

w = Variable(n)
f = F*w
gamma = Parameter(sign='positive')
Lmax = Parameter()
ret = r.T*w 
sigma_risk = quad_form(w, F.T.dot(Sigma).dot(F)+D)
factor_risk = quad_form(f, Sigma) + quad_form(w, D)
prob_factor = Problem(Maximize(ret - gamma*factor_risk), 
               [sum_entries(w) == 1, 
                norm(w, 1) <= Lmax])
prob_sigma = Problem(Maximize(ret - gamma*sigma_risk), 
               [sum_entries(w) == 1, 
                norm(w, 1) <= Lmax])
Lmax.value = 2
gamma.value = 0.1
```

```python
%time
prob_factor.solve(verbose=True)
```
